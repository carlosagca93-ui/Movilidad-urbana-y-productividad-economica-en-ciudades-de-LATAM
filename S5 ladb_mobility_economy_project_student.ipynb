### 3.1 Extraer columna año y filtrar 2024

**🎯Objetivo**
Identificar el año de cada registro y mantener solo los registros del 2024.

**Intrucciones**

- Como el DataFrame `traffic` no tiene una columna de año, utiliza el atributo `.dt.year` sobre su columna de fecha para crear una nueva columna llamada `year`.
- Filtra las filas donde el año sea **2024**.
- Utiliza `.copy()` para crear dos nuevos DataFrames (`traffic_2024` y `eco_2024`) para evitar modificar el dataset original.

<div class="alert alert-block alert-success">

## 🧩Paso 5: Unir movilidad y economía

Combinar datasets te permite analizar cómo se relacionan los indicadores económicos con los de movilidad.

### 5.1 Unir tráfico (tabla principal) con indicadores económicos

**🎯Objetivo:**
Combinar la información de tráfico y economía en un solo DataFrame para analizar cómo las condiciones económicas se relacionan con la movilidad urbana.

**Instrucciones**
- Selecciona solo las **columnas relevantes** de cada dataset (por ejemplo, variables clave de tráfico y de economía).
- Usa `.copy()` al crear subconjuntos para evitar modificar el dataset original.
- Une ambos DataFrames y define como **claves de unión** a `city` y `year`.
- Mantén solo las ciudades y años presentes en ambos datasets.
- Guarda el resultado en una nueva variable llamada `merged` y muestra las primeras 5 filas.


<details>
<summary>Haz clic para ver la pista</summary>
Aplica una unión de tipo "inner" para mantener las ciudades y años presentes en ambos datasets.

```python
# Seleccionar columnas clave de tráfico y economía
left_cols = ['city','country','year','jams_delay','traffic_index_live',
            'jams_length_in_kms','jams_count','mins_delay',
            'travel_time_live_per_10_km_mins','travel_time_historic_per_10_km_mins']

right_cols = ['city', 'year', 'city_gdp/capita', 'unemployment_%', 'pm2.5 (μg/m³)', 'population_(m)']

# Usar .copy() para crear los dos nuevos datasets reducidos
traffic_2024_small = traffic_city_year_2024[left_cols].copy()
eco_2024_small = eco_2024[right_cols].copy()

# Unir datasets
merged = traffic_2024_small.merge(eco_2024_small, on=['city', 'year'], how='inner')

# Mostrar las primeras 5 filas
merged.head()
```

- RESPUESTAS
* Contexto
    La relación entre movilidad urbana y productividad ecónomica es débil y no es lineal cuando se mide por congestion directa.
La variable que mejor predice el PIB per cápita en este conjunto es el desempleo, lo que sugiere que el impacto de la movilidad sobre la productividad se manifiesta más por la vía del mercado laboral, el acceso a empleo, costo de desplazamiento, agotamiento del trabajador, que por el índice de tráfico bruto.

* Cobertura de datos
  Año: 2024
  Ciudades: 15
  Paises: 7

* Metodología (alto nivel)
    Se renombraron las columnas con la finalidad de estandarizar y evitar errores al unir los datasets
    Se corrigen formatos numericos y fecha
    La integración entre trafico y economia  se realizó mediante un INNER JOIN implícito sobre city y year, garantizando que solo registros con datos completos de tráfico y economía estén presentes.
    La distribucion de jams_delay (retrasos por congestion) muestra que México presenta valores anormales (2833) "outlier" superando el limite de (1942), mientras que la media queda en (629.52)

* Hallazgos iniciales
    El patrón dominante es que las ciudades mas congestionadas no son las mas pobres ni las mas ricas, sino las más grandes y activas ("Ciudad de México, Bogotá") concentran los mayores indices de tráfico con poblaciones grandes, aunque ciudades con mejor planificacion urbana tienen menores tiempos de viaje pese a tener un PIB medio.
